#### FEATURE ENGINEERING ON XARRAY

````python
def calculate_daily_averages(wind_data, component):
    # Create lists to store aggregated data for each square
    u_list = []
    v_list = []

    # Extract necessary variables
    longitude = wind_data.longitude.values
    latitude = wind_data.latitude.values

    # Iterate over all combinations of longitude and latitude
    for lon in longitude:
        for lat in latitude:
            # Subset the data for the current square
            square_data = wind_data.sel(latitude=lat, longitude=lon)

            # Determine the variable based on the component parameter
            if component == 'u10':
                component_daily_avg = square_data['u10'].mean(dim='time').item()
                u_list.append(component_daily_avg)
                component_daily_avg = square_data['v10'].mean(dim='time').item()
                v_list.append(component_daily_avg)
            elif component == 'u100':
                component_daily_avg = square_data['u100'].mean(dim='time').item()
                u_list.append(component_daily_avg)
                component_daily_avg = square_data['v100'].mean(dim='time').item()
                v_list.append(component_daily_avg)
            else:
                raise ValueError("Invalid component. Supported values: 'u10', 'v10', 'u100', 'v100'")

    return u_list, v_list
````

#### Calculating Wind speed for Each Squares per Day

````python

def calculate_daily_wind_speed(wind_xarray, component='10'):
    """
    Calculate daily wind speed for each square.

    Parameters:
    - wind_xarray: xarray.Dataset
        Xarray dataset containing wind data.
    - component: str, optional
        Wind component to calculate speed for. Default is '10' (10-meter components),
        can also be '100' (100-meter components).

    Returns:
    - list
        List of daily wind speeds for each square.
    """

    # Create lists to store aggregated data for each square
    wind_speed_list = []

    # Extract longitude and latitude arrays
    longitude = wind_xarray.longitude.values
    latitude = wind_xarray.latitude.values

    # Choose the appropriate wind components
    u_component = f'u{component}'
    v_component = f'v{component}'

    # Iterate over all combinations of longitude and latitude
    for lon in longitude:
        for lat in latitude:
            # Subset the data for the current square
            square_data = wind_xarray.sel(latitude=lat, longitude=lon)

            # Calculate daily averages for the chosen components directly without resampling
            u_daily_avg = square_data[u_component].mean(dim='time').item()
            v_daily_avg = square_data[v_component].mean(dim='time').item()

            # Calculate daily wind speed
            wind_speed_daily = np.sqrt(u_daily_avg**2 + v_daily_avg**2)

            # Append the daily wind speed to the list
            wind_speed_list.append(wind_speed_daily)

    return wind_speed_list


````

#### **calculate_daily_wind_speed Function**

The `calculate_daily_wind_speed` function computes the daily wind speed for each geographic square based on specified wind components (either 10-meter or 100-meter). It takes an xarray dataset containing wind data as input and returns a list of daily wind speeds.
ECMWF recommends we calculate wind speed using: `*Wind_speed = sqrt(u^2+v^2)*`

#### The following code creates a new dataframe for each county and the offshore region showing the average wind speed per day

````python
# Load Kenyan county boundaries from the shapefile
county_shapefile_path = 'data_files/county.shp'
counties_gdf = gpd.read_file(county_shapefile_path)

# Create DataFrame for county-wise wind speed averages
df_county = pd.DataFrame(index=pd.date_range(start='2013-01-01', end='2023-11-30', freq='D'))

# Iterate over all counties
for county_name in counties_gdf['COUNTY']:
    # Find the geometry of the specified county
    county_geom = counties_gdf[counties_gdf['COUNTY'] == county_name].geometry.item()

    # Filter wind speed data within the county geometry
    county_mask = (df['longitude'] >= county_geom.bounds[0]) & (df['longitude'] <= county_geom.bounds[2]) & \
                   (df['latitude'] >= county_geom.bounds[1]) & (df['latitude'] <= county_geom.bounds[3])

    # Group by date and calculate the mean wind speed for each county and day
    county_wind_speed = df.loc[county_mask].groupby(df['time'].dt.date)['wind_speed_daily_100'].mean()

    # Add wind speed data to DataFrame
    df_county[county_name] = county_wind_speed

# Create a mask for offshore coordinates (not in any county and in the ocean)
ocean_mask = (df['latitude'] < -1.60) & (df['longitude'] > 39.3)
offshore_mask = ~df['longitude'].between(counties_gdf.bounds['minx'].min(), counties_gdf.bounds['maxx'].max()) | \
                 ~df['latitude'].between(counties_gdf.bounds['miny'].min(), counties_gdf.bounds['maxy'].max()) | \
                 ocean_mask

# Group by date and calculate the mean wind speed for offshore coordinates
offshore_wind_speed = df.loc[offshore_mask].groupby(df['time'].dt.date)['wind_speed_daily_100'].mean()

# Set the values in the 'Offshore' column
df_county['Offshore'] = offshore_wind_speed

# Print or use the DataFrame as needed
df_county.head(5)

````

#### Generate Quiver Plot Function

````python


def generate_quiver_plot(wind_xarray, u_list, v_list, wind_speed_list, component='100m', cmap='viridis'):
    # Your longitude, latitude, and wind speed data
    longitude = wind_xarray.longitude.values
    latitude = wind_xarray.latitude.values

    # Specify the area covered by the 1287 squares
    north = 5.0
    west = -33.9
    south = -4.5
    east = 41.9

    # Filter data within the specified area
    mask = (latitude >= south) & (latitude <= north)
    filtered_longitude = longitude[(longitude >= west) & (longitude <= east)]
    filtered_latitude = latitude[mask]

    # Create a meshgrid for the quiver plot
    lon_grid, lat_grid = np.meshgrid(filtered_longitude, filtered_latitude, indexing='ij')
    u_grid = np.array(u_list).reshape(lon_grid.shape)
    v_grid = np.array(v_list).reshape(lon_grid.shape)
    wind_speed_grid = np.array(wind_speed_list).reshape(lon_grid.shape)

    # Create a quiver plot
    fig, ax = plt.subplots(figsize=(10, 8), subplot_kw=dict(projection=ccrs.PlateCarree()))
    quiver = ax.quiver(lon_grid, lat_grid, u_grid, v_grid, wind_speed_grid, cmap=cmap, scale=50, width=0.003)

    # Add coastlines and international boundaries
    ax.coastlines(resolution='10m', linewidth=1)
    ax.add_feature(NaturalEarthFeature(category='cultural', name='admin_0_boundary_lines_land', scale='10m', edgecolor='black', facecolor='none'))

    # Create a separate colorbar for the wind speed
    sm = ScalarMappable(cmap=cmap)
    sm.set_array(wind_speed_list)
    cbar = plt.colorbar(sm, ax=ax, label='Wind Speed (m/s)')

    ax.set_title(f'Wind Direction and Speed ({component})')

    plt.xlabel('Longitude')
    plt.ylabel('Latitude')

    plt.show()

````



The `generate_quiver_plot` function creates a visually appealing quiver plot to represent wind direction and speed over a specified geographical area. It utilizes data from an xarray dataset.

### Parameters:

- `wind_xarray`: Xarray dataset containing wind data.
- `u_list`: List of U-component wind data.
- `v_list`: List of V-component wind data.
- `wind_speed_list`: List of corresponding wind speeds.
- `component` (optional): Wind component, either '10m' or '100m'.
- `cmap` (optional): Colormap for the quiver plot.

### INDUSTRIAL POWER PRODUCTION RATE

````python

def industrial_turbine_power_curve(wind_speed, cut_in_speed=2.0, rated_speed=15.5, cut_off_speed=25.0, rated_power=2300.0):
    # Power curve model
    power_curve = np.zeros_like(wind_speed)
    
    # Below cut-in speed, power is zero
    power_curve[wind_speed < cut_in_speed] = 0.0
    
    # Between cut-in speed and rated speed, use a cubic function
    power_curve[(wind_speed >= cut_in_speed) & (wind_speed <= rated_speed)] = (
        rated_power * ((wind_speed[(wind_speed >= cut_in_speed) & (wind_speed <= rated_speed)] - cut_in_speed) / (rated_speed - cut_in_speed))**3
    )
    
    # Above rated speed and up to cut-off speed, power is constant at rated power
    power_curve[(wind_speed > rated_speed) & (wind_speed <= cut_off_speed)] = rated_power
    
    # Above cut-off speed, power is zero
    power_curve[wind_speed > cut_off_speed] = 0.0

    return power_curve
````

#### **POWER OUTPUT CALCUTATION FUNCTION**

````python
def calculate_power_output(df, cut_in_speed, rated_speed, cut_off_speed, rated_power, time_period_hours, lon_grid, lat_grid, county_shapefile_path):
    # Function to calculate power output based on turbine power curve
    def industrial_turbine_power_curve(wind_speed, cut_in_speed, rated_speed, cut_off_speed, rated_power, time_period_hours):
        power_output = np.zeros_like(wind_speed)
        mask = (wind_speed >= cut_in_speed) & (wind_speed <= cut_off_speed)
        power_output[mask] = (wind_speed[mask] - cut_in_speed) / (rated_speed - cut_in_speed) * rated_power * time_period_hours / 1000  # Convert to MWh
        mask = wind_speed > cut_off_speed
        power_output[mask] = rated_power * time_period_hours / 1000  # Convert to MWh
        return power_output

    # Group by time period and calculate the mean wind speed
    if time_period_hours == 1:
        grouped_data = df.groupby(df['time'].dt.to_period("H")).agg({'wind_speed_daily_100': 'mean'})
    elif time_period_hours == 24:
        grouped_data = df.groupby(df['time'].dt.to_period("D")).agg({'wind_speed_daily_100': 'mean'})
    elif time_period_hours == 168:
        grouped_data = df.groupby(df['time'].dt.to_period("W")).agg({'wind_speed_daily_100': 'mean'})
    elif time_period_hours == 730:  # Assuming 30 days in a month
        grouped_data = df.groupby(df['time'].dt.to_period("M")).agg({'wind_speed_daily_100': 'mean'})
    elif time_period_hours == 8760:  # Assuming 365 days in a year
        grouped_data = df.groupby(df['time'].dt.to_period("Y")).agg({'wind_speed_daily_100': 'mean'})
    else:
        raise ValueError("Unsupported time period")

    # Apply turbine power curve to estimate power output for each time period
    grouped_data['estimated_power_output_MWh'] = industrial_turbine_power_curve(grouped_data['wind_speed_daily_100'], 
                                                                                   cut_in_speed, rated_speed, cut_off_speed, rated_power, time_period_hours)

    # Load Kenyan county boundaries from the shapefile
    counties_gdf = gpd.read_file(county_shapefile_path)

    # Create a GeoDataFrame from the wind speed data
    geometry = [Point(lon, lat) for lon, lat in zip(lon_grid.flatten(), lat_grid.flatten())]
    wind_speed_gdf = gpd.GeoDataFrame({'geometry': geometry, 'wind_speed': wind_speed_grid.flatten()})

    # Add county information to the wind speed GeoDataFrame
    for index, county in counties_gdf.iterrows():
        mask = wind_speed_gdf.within(county['geometry'])
        wind_speed_gdf.loc[mask, 'county'] = county['COUNTY']

    # Create a list to store results
    result_data = []

    # Calculate power output and aggregate for each county on the specified time period
    for county_name, county_data in tqdm(wind_speed_gdf.groupby('county')):
        county_data['power_output'] = industrial_turbine_power_curve(county_data['wind_speed'], cut_in_speed, rated_speed, cut_off_speed, rated_power, time_period_hours)
        total_power_output = county_data['power_output'].sum()
        average_power_output = total_power_output / len(county_data)
        average_wind_speed = county_data['wind_speed'].mean()  # Calculate average wind speed

        # Append results to the list
        result_data.append({
            'county': county_name,
            'average_wind_speed': average_wind_speed,
            f'total_power_output_{time_period_hours}h_MWh': total_power_output,
            f'average_power_output_{time_period_hours}h_MWh': average_power_output,
            # 'average_wind_speed': average_wind_speed
        })

    # Create a DataFrame from the list of results
    result_df = pd.DataFrame(result_data)

    # Replace 'nan' with 'Offshore' in the 'county' column
    result_df.loc[result_df['county'] == 'nan', 'county'] = 'Offshore'

    return result_df
````

#### **Power Output Calculation Function**

The provided code defines a function, `calculate_power_output`, which computes the estimated power output from wind data. The function takes a DataFrame (`df`) containing wind speed data, along with parameters such as cut-in speed, rated speed, cut-off speed, rated power, and time period in hours. It utilizes an industrial turbine power curve to estimate power output for specified time intervals.

### Steps in the Function:
1. **Wind Speed Grouping:**
   - The function groups the wind speed data by the specified time periods (e.g., hourly, daily, weekly, monthly, yearly).

2. **Power Output Estimation:**
   - It applies the turbine power curve to estimate power output for each time period.

3. **Spatial Analysis:**
   - The function loads Kenyan county boundaries from a shapefile and associates wind speed data with respective counties.

4. **County-Level Aggregation:**
   - It calculates power output for each county, considering average wind speed, total power output, and average power output for the specified time period.

5. **Result Presentation:**
   - The function returns a DataFrame summarizing the calculated power output for each county.

*Note: The function uses the tqdm library for progress tracking during county-level computations.*

### **RECOMMENDATIONS**

##### 1. **CHOICE OF THE WIND TURBINES**
- ##### Vesta52 850kw Turbines which have 850kwh. 
- These turbines are overwhelmingly used in the country. 
- While these turbines are cheaper compared to other models in the industry, they also produce dismally low power output (The entire Lake Turkana Wind Farm has 165 Turbines and only produces 310MW)
- These Turbines also have a high Cut-in-speed(Speed at which the turbine starts operating) (4.0m/s) which is very high. 
This is a prohibitive in terms of applications to many areas.
- These turbines also have high optimal operating speed ( speed at which the turbine is fully efficient ad produces maximum power) (15m/s). Very few places have this kind of wind speed for reasonably long time. Even the high wind speed areas hardly reach those speeds. they top at 12.5m/s


##### Siemens SWT 3.6 - 120T
- This turbine has higher capacity (3.6 MW at full capacity)
- They are also have very low cut-in-speed (3.0m/s) making them more suitable for more places
- They also have a low rated speed (12.5m/s) which makes it reasonably well suited to many areas as well
- If Lake turkana replaced the 165 Vesta52 turbines with this one, it would theoretically supply Kenya's Energy need only 11 turbines (each producing roughly 29 mwh a day when operating at 34% operating efficiency)
- One downside of this turbine is prohibitively costly (at $2.4 million per Unit)
- It is important to know that these turbines have a 10-12 years to recoup their purchase cost (in the case of Turkana): 
It is important to know that average lifespan of these wind turbines is 25 years.

`Therefore, Switching to the high performing and higher output turbine is a reasonably good option notwithstanding the high installation cost.`
